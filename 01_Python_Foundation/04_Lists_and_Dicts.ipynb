Single variables (`a = 1`) are rare in Data Engineering. We deal with **Datasets**.
We need structures to hold **collections** of data.

1. **List `[]`**: Ordered sequence (e.g., list of filenames).
2. **Dictionary `{}`**: Key-Value lookup (e.g., configuration, JSON).
3. **Set `{}`**: Unique values (e.g., deduplicating IDs).

---

## 1. Lists (Arrays)
Ordered. Mutable (can change). duplicates allowed.

```python
files = ["data.csv", "style.css", "script.py"]

# Accessing items (0-indexed)
print(files[0])  # First file
print(files[-1]) # Last file

# Adding items
files.append("readme.md")
print(files)
```

## 2. Dictionaries (Hash Maps)
The **most important** structure for Data Engineers. It's how we handle JSON data, configurations, and API responses.
Think of it like a real dictionary: Look up a word (Key) to get the definition (Value).

```python
azure_config = {
    "resource_group": "rg-data-prod",
    "location": "eastus",
    "vm_size": "Standard_D2s_v3",
    "active": True
}

# Lookups are FAST (O(1))
print(azure_config["location"])

# safely getting a value that might missing
print(azure_config.get("tags", "No Tags"))
```

## 3. Nested Structures (JSON-like)
In the real world, data is nested. A list of dictionaries.

```python
users = [
    {"id": 1, "name": "Alice", "role": "Admin"},
    {"id": 2, "name": "Bob", "role": "Viewer"}
]

for user in users:
    print(f"User {user['name']} is a {user['role']}")
```
