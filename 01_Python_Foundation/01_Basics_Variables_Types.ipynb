## 2. Variables (Containers)
A variable is like a box where you store data to use later.
You give the box a name, and put a value inside using `=`.

```python
user_name = "Sarah"
age = 25

print(user_name)
print(age)
```

## 3. Data Types
Data comes in different "flavors". Python automatically detects the type, but you must know them.

1. **String (`str`)**: Text. Wrapped in quotes `'` or `"`. E.g., `"Azure"`
2. **Integer (`int`)**: Whole numbers. E.g., `10`, `42`, `-5`
3. **Float (`float`)**: Decimal numbers. E.g., `10.5`, `3.14`
4. **Boolean (`bool`)**: Truth values. Only `True` or `False`.

```python
company = "Microsoft"  # str
year = 2024            # int
price = 99.99          # float
is_active = True       # bool

# checking types
print(type(company))
print(type(price))
```

## 4. Basic Math
Python is a calculator.

```python
a = 10
b = 3

print(a + b)  # Addition (13)
print(a - b)  # Subtraction (7)
print(a * b)  # Multiplication (30)
print(a / b)  # Division (3.333...)
print(a // b) # Floor Division (3) -> Removes decimal
print(a % b)  # Modulo (1) -> Remainder of 10 / 3
```

## 5. f-strings (String Formatting)
The modern way to combine text and variables. Put an `f` before the quotes and use `{}` for variables.

```python
name = "Data Factory"
pipelines = 5

# New Way (Best)
message = f"Count of {name} pipelines: {pipelines}"
print(message)
```
