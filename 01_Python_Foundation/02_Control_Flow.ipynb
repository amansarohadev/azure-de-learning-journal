## 1. Logic: The `if` Statement
We use `if` to run code ONLY when a condition is True.

**Operators:**
- `==` (Equal to)
- `!=` (Not equal to)
- `>` (Greater than)
- `<` (Less than)
- `>=` (Greater or Equal)
- `<=` (Less or Equal)

```python
temperature = 35

if temperature > 30:
    print("It's hot outside! â˜€ï¸")
    
print("Done checking.")
```

### `else` and `elif` (Else If)
To handle multiple cases.

```python
score = 85

if score >= 90:
    print("Grade: A ðŸŒŸ")
elif score >= 80:
    print("Grade: B ðŸ‘")
else:
    print("Grade: C")
```

## 2. Logical Operators (`and`, `or`, `not`)
Combine multiple conditions.

```python
is_weekend = False
has_coffee = True

if is_weekend or has_coffee:
    print("I am happy. â˜•")
    
if not is_weekend:
    print("Go to work. ðŸ’¼")
```

## 3. Loops: Repeating Actions
In Data Engineering, you rarely process 1 file. You process **1,000** files. Loops let you write code once and run it many times.

### The `for` Loop + `range()`
`range(5)` creates a sequence: 0, 1, 2, 3, 4.

```python
print("Starting Countdown...")

for number in range(5):
    print(f"Processing item #{number}")
    
print("Finished!")
```
