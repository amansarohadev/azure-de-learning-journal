Scripts get messy. **Classes** organize code into logical units.
In Data Engineering, we use classes to build **Connectors**, **Pipelines**, and **Models**.

---

## 1. Class vs Object
- **Class**: The Blueprint (e.g., "Car Design").
- **Object**: The actual thing built from the blueprint (e.g., "My Red Toyota").

```python
class DataPipeline:
    # The Constructor (__init__): Runs when you create the object
    def __init__(self, name, source):
        self.name = name
        self.source = source
        print(f"Pipeline '{self.name}' initialized from {self.source}.")

    # A Method (Function inside a class)
    def run(self):
        print(f"Running pipeline: {self.name}...")

# Creating Objects (Instances)
p1 = DataPipeline("SalesETL", "SQL Server")
p2 = DataPipeline("LogsETL", "Blob Storage")

p1.run()
p2.run()
```

## 2. Inheritance (Don't Repeat Logic)
You can create a specialized version of a class.

```python
class SqlPipeline(DataPipeline):
    def connect(self):
        print(f"Connecting to SQL Database at {self.source}...")

# Inherits everything from DataPipeline
sql_job = SqlPipeline("DailySales", "prod-db-01")
sql_job.run()      # From Parent
sql_job.connect()  # Specific to SqlPipeline
```

### Task 1: The AzureBlobConnector
1. Create a class `AzureBlobConnector`.
