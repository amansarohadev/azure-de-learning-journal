## 1. Arrays vs Lists
NumPy arrays look like lists, but they are:

```python
import numpy as np

# A standard python list
py_list = [1, 2, 3]

# A numpy array
np_arr = np.array([1, 2, 3])

print(f"List: {py_list}")
print(f"Array: {np_arr}")
```

## 2. Vectorization (The Magic)
If you want to multiply every number by 2:
- **List**: You need a loop (Slow).
- **NumPy**: You just do `array * 2` (Super Fast).

```python
# The Old Way (Loop)
new_list = []
for x in py_list:
    new_list.append(x * 2)

# The NumPy Way (Vectorized)
new_arr = np_arr * 2

print(f"Loop Result: {new_list}")
print(f"NumPy Result: {new_arr}")
```

## 3. Filtering (Masking)
Select data without `if` statements.

```python
data = np.array([10, 50, 80, 20, 90])

# Create a mask (True/False)
high_value_mask = data > 40
print(f"Mask: {high_value_mask}")

# Apply the mask
filtered_data = data[high_value_mask]
print(f"Filtered: {filtered_data}")
```
