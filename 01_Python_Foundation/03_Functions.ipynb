## 1. Defining a Function
Use `def` to create a function.
The code inside does NOT run until you **call** these function.

```python
def say_hello():
    print("Hello from inside the function! ðŸ‘‹")

# Calling it
say_hello()
say_hello()
```

## 2. Parameters (Inputs)
Functions become powerful when they can accept data.

```python
def greet_user(name):
    print(f"Welcome back, {name}!")

greet_user("Alice")
greet_user("Bob")
```

## 3. Return Values (Outputs)
Most Data Eng functions don't just print; they **calculate** something and give it back.
Use `return`.

```python
def calculate_tax(price):
    tax_rate = 0.10
    return price * tax_rate

my_price = 100
tax = calculate_tax(my_price)

print(f"Price: {my_price}, Tax: {tax}, Total: {my_price + tax}")
```

### Why `return`? 
If you use `print` inside a function, you can't use the result in another calculation.
`return` hands the value back to the main program.
