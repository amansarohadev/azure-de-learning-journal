## The Challenge

Create a class `ETLPipeline` with the following methods:

1.  `__init__(self, input_file, output_file)`: Store file paths.
2.  `extract(self)`: Read the CSV. 
    -   **Constraint**: Use `try/except` to catch `FileNotFoundError`.
    -   Return the DataFrame.
3.  `transform(self, df)`:
    -   Add a column `total_amount` (`price * quantity`).
    -   Filter **OUT** any rows where `status` is "Return".
    -   Group by `product` and calculate the `sum` of `total_amount`.
    -   Return the summarized DataFrame.
4.  `load(self, df)`:
    -   Save the summarized DataFrame to `self.output_file` as JSON (`df.to_json()`).
5.  `run(self)`: Orchestrate the steps (Extract -> Transform -> Load).

**Goal Output (`sales_summary.json`):**
It should show total sales per product (excluding returns).

```python
# TODO: Write your Class here
class ETLPipeline:
    import pandas as pd
    def __init__(self,input_file,output_file):
        self.input_file = input_file
        self.output_file = output_file
    def extract(self, input_file):
        try:
            df = pd.read_csv(input_file)
            return df
        except Exception as e:
            print(f"Error reading file: {e}")
            return None
    def transform(self, df):
        try:
            df['total_amount'] = df['price'] * df['quantity']
            df=df[df['status']!='Return'] 
            return   df.groupby('product')['total_amount'].sum().reset_index()
        except Exception as e:
            print(f"Error transforming data: {e}")
            return None
    def load(self, df):
        try:
            df.to_json("sales_summary.json", orient="records")
            print(f"Data loaded to {self.output_file}")
        except Exception as e:
            print(f"Error loading data: {e}")
    def run(self):
        df = self.extract(self.input_file)
        if df is not None:
            df = self.transform(df)
            if df is not None:
                self.load(df)
```

```python
# TODO: Run the pipeline
pipeline = ETLPipeline("raw_orders.csv", "sales_summary.json")
pipeline.run()
```
